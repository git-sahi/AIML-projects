```python
# ✅ STEP 1: Load Documents
from langchain_community.document_loaders import WebBaseLoader

loader=WebBaseLoader("https://www.pregrad.in")
docs=loader.load()
```

```python
# ✅ STEP 2: Split the Text
from langchain.text_splitter import RecursiveCharacterTextSplitter

splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
chunks = splitter.split_documents(docs)
```

```python
len(chunks)
```

```python
# ✅ STEP 3: Create Embeddings
api_key="AIzaSyDdQlEYR4DDfSKWXDC8rL***" #replace with
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings

embedding_model = GoogleGenerativeAIEmbeddings(model="models/text-embedding-004",google_api_key=api_key)
```

```python
text= "we are langchan"
```

```python
result=embedding_model.embed_query(text)
```

```python
# ✅ STEP 4: Store in Vector DB
from langchain.vectorstores import FAISS

vectorstore = FAISS.from_documents(chunks, embedding_model)
print("✅ Vector store ready!")
```

```python
retriver=vectorstore.as_retriever()
query="what is ai"
```

```python
retriver.invoke(query)
```

```python
# ✅ STEP 5: Ask Questions with RAG
from langchain_google_genai import ChatGoogleGenerativeAI
llm =ChatGoogleGenerativeAI(model="gemini-2.0-flash",api_key=api_key)
```

```python
from langchain.chains import RetrievalQA
qa_chain = RetrievalQA.from_chain_type(llm=llm, retriever=vectorstore.as_retriever())
```

```python
result= llm.invoke("what is ai?")
```

```python
query = "What is pregrad?"
response = qa_chain.run(query)

print("🤖 Answer:", response)
```
